## Condicional

```python
# If
if 5 < 10:
  print('Verdade')
```

```python
# If/else
num1 = 10
num2 = 5
if num1 > num2:
  print('O num1 é maior que o num2')
else:
  print('O num2 é maior que o num1')
```

```python
# If/Elif/Else
num1 = 10
num2 = 10
if num1 > num2:
  print('O num1 é maior que o num2')
elif num1 < num2:
  print('O num1 é menor que num2')
else:
  print('Os números são iguais')
```

## Condicional com Operadores Lógicos

```python
num1 = 10
num2 = 5
num3 = 2
if (num1 > num2) and (num1 > num3):
  print('O num1 é o maior')
else:
    print('O num1 não é o maior')
```
